CELL 10: Kiểm tra các giá trị lạ trong tất cả các cột trước khi xử lý

```python
special_chars = ['?', '...', '\t']

for col in df.columns:
    unique_vals = df[col].dropna().unique()
    strange_vals = []

    for val in unique_vals:
        val_str = str(val)
        if val_str.strip() != val_str or any(ch in val_str for ch in special_chars):
            strange_vals.append(val)

    if strange_vals:
        print(f"\nCột '{col}' có giá trị lạ:")
        print(strange_vals)
```

CELL 17: Xử lý giá trị thiếu mỗi cột

```python
print("\n=== Xử lý missing values ===")

for col in numeric_cols:
    missing = df[col].isnull().sum()
    if missing > 0:
        median_val = df[col].median()
        df[col] = df[col].fillna(median_val)
        print(f"Cột '{col}': điền {missing} missing bằng median = {median_val}")

for col in categorical_cols:
    df[col] = df[col].replace(['nan', 'NaN', ''], np.nan)
    missing = df[col].isnull().sum()
    if missing > 0:
        mode_val = df[col].mode()[0]
        df[col] = df[col].fillna(mode_val)
        print(f"Cột '{col}': điền {missing} missing bằng mode = '{mode_val}'")
```

CELL 21: Kiểm tra các outlier (theo phương pháp IQR) trước khi xử lý

```python
outlier_summary = []


for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()

    if outlier_count > 0:
        outlier_summary.append([col, outlier_count])
        print(f"Cột '{col}' có {outlier_count} giá trị outlier")
```

CELL 23: Xử lý các outlier bằng IQR capping

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()

    if outlier_count > 0:
        df[col] = np.where(df[col] < lower, lower, df[col])
        df[col] = np.where(df[col] > upper, upper, df[col])

        print(f"Cột '{col}': đã xử lý {outlier_count} outlier (capping [{lower:.2f}, {upper:.2f}])")
```

CELL 24: Kiểm tra lại các outlier sau xử lý (IQR capping)

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outlier_count = ((df[col] < lower) | (df[col] > upper)).sum()

    if outlier_count > 0:
        print(f"Cột '{col}' vẫn còn {outlier_count} outlier")
```

CELL 27: Mã hoá biến phân loại (Categorical Encoding)

```python
binary_map = {
    'yes': 1, 'no': 0,
    'normal': 1, 'abnormal': 0,
    'good': 1, 'poor': 0,
    'present': 1, 'notpresent': 0
}

binary_cols = [
    'red_blood_cells',
    'pus_cell',
    'pus_cell_clumps',
    'bacteria',
    'hypertension',
    'diabetes_mellitus',
    'coronary_artery_disease',
    'appetite',
    'pedal_edema',
    'anemia'
]

for col in binary_cols:
    df[col] = df[col].map(binary_map)

df[binary_cols] = df[binary_cols].fillna(0).astype(float)

df['classification'] = df['classification'].map({
    'ckd': 1,
    'notckd': 0
})

df = df.dropna(subset=['classification'])
df['classification'] = df['classification'].astype(float)
```
